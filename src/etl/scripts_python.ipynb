## funciones at y loc

```python
df_copy = df.copy()
df_copy.at[0,'sub_region_name'] = "Latino america y el caribe"
df_copy.head(2)
```

```python
for indice, fila in  df_copy.iterrows():
    val = df_copy.at[indice,'sub_region_name']
    if val == "Latin America and the Caribbean":
        df_copy.at[indice,'sub_region_name'] = "Latino america y el caribe"
df_copy.head(2)
```

```python
df_copy.loc[df_copy['sub_region_name'] == 'Latino america y el caribe']
```

```python
df_copy.groupby('sub_region_name')['sub_region_name']
df_copy.groupby('sub_region_name')['sub_region_name'].head()
```

```python
filtered_df = df_copy[df_copy['sub_region_name'] == 'Latino america y el caribe']
filtered_df
```
